```python
data=pd.read_csv("/content/grive/MyDrive/Tensorflow-demo/birds_data.csv", header=None)
```

```python
data.describe()
```

```python
# Convert data to array
data = data.to_numpy()
```

```python
# Print array shape
data.shape
```

```python
# Load training and testing data
training = data[0:60]
testing = data[60:]
```

```python
print(training)
print(testing)
```

```python
training_features=training[:,0:-1]
training_labels=training[:,-1]
testing_features=testing[:,0:-1]
testing_labels=testing[:,-1]
```

```python
print(training_features.shape)
print(training_labels.shape)
```

```python
training_features[0].shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
# Model definition
model=Sequential()
model.add(Input(shape=(20,)))
model.add(Dense(18,activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(1,activation='tanh'))
```

```python
# Model compile
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
print(training_features.shape)
print(training_labels.shape)
print(testing_features.shape)
print(testing_labels.shape)
```

```python
model.fit(training_features,training_labels,epochs=1000,validation_data=(testing_features,testing_labels))
```
